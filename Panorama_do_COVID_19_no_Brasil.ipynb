* Para identificação os paises que tem mais numero de casos vai ser feito o tratamento dos dados e depois identificado quando foi feito a informação mais recente.

```
#Criando uma copia para manipulação
df_loc = df.copy()

# Filtrar os registros que não são "NaN" na coluna "continent" e "total_deaths"
df_filtered = df_loc.dropna(subset=['continent', 'total_deaths'])

# Identificar a data mais atual depois do tratamento
data_filered_max = df_filtered.date.max()

# Filtrar os registros onde a data na coluna 'date' é igual à data máxima
df_final = df_filtered[df_filtered['date'] == data_filered_max]

# Organizar o DataFrame 'df' pelo total de mortes em ordem decrescente
df_sorted = df_final.sort_values(by='total_deaths', ascending=False)

# Pegar os 5 primeiros da coluna "location"
locations = df_sorted.head(5)[['location', 'total_deaths']]

# Definir a coluna 'location' como índice
locations.set_index('location', inplace=True)

print('A data mais recente dos dados por paises é : ', data_filered_max.day,'/',data_filered_max.month,'/',data_filered_max.year)
print('\n')
print(locations)
```

```
# Selecionando os países desejados e calculando o total de mortes em cada um deles
paises_selecionados = locations.index
total_de_mortes_por_pais = df.loc[df['location'].isin(paises_selecionados)].groupby('location')['total_deaths'].max()
# Ordenar do maior para o menor
total_de_mortes_por_pais = total_de_mortes_por_pais.sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))

total_de_mortes_por_pais.plot(kind='bar', ax=ax, color='r')

ax.set_title("5 paises com mais mortes")
ax.set_xlabel("Paises")
ax.set_ylabel("Total de mortes por mi")


plt.tight_layout()
```

####Heatmap:

* Isso permite visualizar padrões e relações entre as variáveis de forma mais clara e intuitiva do que simplesmente examinar os números brutos.

* Interpretar o heatmap, Analise a saída para identificar padrões, tendências e correlações entre as variáveis. Isso pode ajudar a gerar insights sobre os dados e a informar análises e decisões subsequentes.

```
# Selecionando as colunas relevantes do DataFrame
colunas_selecionadas = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths','reproduction_rate',
                        'total_tests', 'new_tests','total_vaccinations', 'people_vaccinated',
                        'people_fully_vaccinated', 'total_boosters', 'new_vaccinations',
                        'new_people_vaccinated_smoothed', 'stringency_index', 'excess_mortality_cumulative_absolute',
                        'excess_mortality_cumulative','excess_mortality']

df_heatmap = df_bra[colunas_selecionadas]

# Calculando a matriz de correlação
correlation_matrix = df_heatmap.corr()

# Criando uma máscara para esconder a metade superior
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Plotando o heatmap
plt.figure(figsize=(24, 20))
sns.heatmap(correlation_matrix, mask=mask ,annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Heatmap de Correlação entre Variáveis')

# Exibindo o plot
plt.show()
```

* Para pegar os itens que têm correlação acima de 85%, pode iterar sobre a matriz de correlação e selecionar os pares de variáveis que atendem a esse critério. Assim a analise de correlação entre grandes bases fica mais fácil de ser lida e interpretada.

* Este código iterará sobre a matriz de correlação e verificará se o valor absoluto da correlação entre duas variáveis é maior que o limite especificado. Se for, o par de variáveis e a correlação serão adicionados à lista pares_correlacao_alta. Em seguida, o código imprimirá os pares de variáveis com correlação acima do limite especificado.

```
# Definindo um limite de correlação
limite_correlacao = 0.85

# Criando uma lista para armazenar os pares de variáveis com correlação acima do limite
pares_correlacao_alta = []

# Iterando sobre a matriz de correlação
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > limite_correlacao:
            # Adicionando o par de variáveis à lista
            pares_correlacao_alta.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))

# Convertendo a lista de pares de correlação alta para um DataFrame do pandas
df_pares_correlacao_alta = pd.DataFrame(pares_correlacao_alta, columns=['Variável 1', 'Variável 2', 'Correlação'])

# Salvando os dados em um arquivo Excel
df_pares_correlacao_alta.to_excel('pares_correlacao_alta.xlsx', index=False)

# Exibindo uma mensagem de confirmação
print("Dados exportados com sucesso para o arquivo 'pares_correlacao_alta.xlsx'.")
```
